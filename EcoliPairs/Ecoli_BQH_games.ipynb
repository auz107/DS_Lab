###The region, where (m1,m2) + (m1_m2, wild_type) are the Nash equilibria
**When WT invades**
   - WIld-type will have a very higher fraciton (e.g., 95%). This is because wild-type strais can always survive no matter who may meet (they can synthesize all what they need)
   - m1 and m2 coexist in a small fraction (e.g., each 2%). This is because, m1 can surive only when meeting m2 or a wild-type and vice versa. Since there are too many of these mutants initially, they can co-exist. Increasing the leakiness levels increases the percentage of these mutants by only a small value (e.g., to 3%). 
   - The double mutant m1_m2 is driven to extinciton. This is because m1_m2 can survive only if it meets a wild-type. Initially, there is only a small fraction of wild-type straiins, so they go extinct before wild-type strains dominate
    
    
**When mutants invade**
   - WIld-type will have a very higher fraciton (e.g., 95%). This is because wild-type strais can always survive no matter who may meet (they can synthesize all what they need)
   - m1 and m2 go extinct. This is because, m1 can surive only when meeting m2 or a wild-type and vice versa. Since the initial fraction of these mutants is very small, they can co-exist in only small fractions. Additionally, note that m1 or m2 facing a wild-type are less fit than an m1_m2 facing the wild-type, so they cano compete with m1_m2
   - The double mutant m1_m2 co-exists with wild-type but in small fractions (e.g., 5%), however, this fraction increases (e.g., to 30%) as the the leakiiness levels increase. This is because there are initally too many wild-type strains and it's easy for m1_m2 mutants to wfind a wild-type partner to rely on


###Lower right region where in addition to Prisoner's Dilemmas we have a Nash equilibrium (m1,m1_m2):

In this region, that high level of leaky trait 1 is not possible. Therefore, any strain that has to produce the leaky trait (i.e., WT and m2) will go extinct but the rest (i.e., m1 and m1_m2) can survie. However, even though when m1 and m1_m2 face each other they don't get any benefits (i.e., a payoff of zero) but they will be more fit than other strains that need to produce those leaky trait 1 (leading to a negative payoff). This pattern holds for both case of wild-type invading and mutants invading. 

**When WT invades**
   - The fraction of m1 is higher than m1_m2 because initially we have a small fraciton of wild-type strains and m1_m2 cannot benefit from wild-type 
    
**When mutants invade**
   - The fraction of m1_m2 is higher than that for m1 because the inital high fracitno of wild-type strains allow m1_m2 to get more benefits
   

###Top left region where in addition to Prisoner's Dilemmas we have a Nash equilibrium (m2,m1_m2):

All patterns here are the reverse of what mentioned for the lower right region

###Top right corner region where Nash equilibria are Prisoner's Dilemmas (m1,m1), (m1,m2), (m1_m2, m1_m2):

In this region no strain can produce either leaky trait 1 or leaky trait 2 at that high levels, Therefore, m1_m2 will be the fittest strain as it doesn't have t produce any of these leaky trais and it dominates the co-culture.
